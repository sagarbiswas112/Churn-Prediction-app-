## Installing Dependencies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Phase 1: Data Cleaning Process

```python
df = pd.read_csv('D:/project/Telco-Customer-Churn.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df.drop('customerID' , axis = 1, inplace = True)
```

#### axis=1: This is a crucial argument. It specifies what we are dropping.
 * axis=0 refers to rows.
 * axis=1 refers to columns.
#### inplace=True: This argument controls how Pandas makes the change.
 * If inplace=False (the default), Pandas creates a new copy of the DataFrame with the column removed and leaves your original df untouched.
 * By setting inplace=True, we are telling Pandas: "Don't make a copy. Modify my original df directly." This is more memory-efficient and common for data cleaning.

```python
print(df.head())
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
```

* pd.to_numeric(...): This is a powerful Pandas function designed specifically for this task. It attempts to convert whatever you give it into a number.
* errors='coerce': This is the most important part of the command. It's an instruction for what to do if it finds a value that it cannot convert into a number (like our blank spaces).
* The word coerce means to force or pressure.
By setting errors='coerce', we are telling Pandas: "If you encounter a value you can't turn into a number, don't stop and give me an error. Instead, force it to become NaN."
NaN stands for Not a Number, and it's the standard way Pandas marks missing data.

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace= True)
```

```python
print(df.info())
```

## Phase 2: Feature Engineering and Preprocessing

#### Separate Features (X) from the Target (y)
 * Features (X): The set of all input columns we will use to make a prediction (e.g., tenure, Contract, MonthlyCharges).
 * Target (y): The single column we are trying to predict (in our case, Churn).

```python
x = df.drop('Churn', axis = 1)
y = df['Churn']
```

#### Encode the Target Variable (y)

```python
y = y.map({'Yes': 1 ,'No': 0})
```

* The .map() method is a powerful way to substitute each value in a Series with another value.

#### Performing One-Hot Encoding on x (Features)

```python
x_final = pd.get_dummies(x, drop_first = True)
```

* pd.get_dummies(...): This is the star of the show. It's the Pandas function that performs one-hot encoding.
* drop_first=True: This is an important argument for efficiency and statistical best practices. Let's look at the gender column, which has 'Male' and 'Female'. The function will create a new column, gender_Male. If a customer's value in gender_Male is 1, we know they are male. If the value is 0, we automatically know they must be female. We don't need a separate gender_Female column because it would be redundant. drop_first=True tells the function to drop one category from each feature to avoid this redundancy.

```python
print(x_final.head())
```

```python
print(x_final.columns)
```

```python
from sklearn.model_selection import train_test_split
```

* from sklearn.model_selection: We are going into the scikit-learn library (sklearn) and accessing its model_selection module, which contains tools for preparing data and evaluating models

```python
x_train,x_test, y_train, y_test = train_test_split(x_final, y , test_size = 0.2, random_state = 42)
```

* The train_test_split function returns four dataframes, and we are assigning them to these four new variables in this specific order.
* test_size=0.2: This argument specifies the proportion of the data to be allocated to the test set. 0.2 means we will use 20% of the data for testing and the remaining 80% for training.
* random_state=42: The function shuffles the data randomly before splitting. Setting a random_state to any number (42 is a popular convention) ensures that the shuffle is the exact same every time you run the code. This is crucial for reproducibility, so you and others can get the same results.

```python
print('x_train_shape', x_train.shape)
print('x_test_shape' , x_test.shape)
```

* .shape: This is an attribute (not a method, so no ()) of every dataframe that returns its dimensions as (rows, columns).

### Scale the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_trained_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

* scaler = StandardScaler(): We create an instance of the scaler object.
* X_train_scaled = scaler.fit_transform(X_train):
* fit_transform(X_train): This is a two-in-one command. It first learns the scaling parameters (the mean and standard deviation) from X_train (fit) and then applies the transformation to X_train (transform).
* X_test_scaled = scaler.transform(X_test):
* transform(X_test): Notice we only use transform here, not fit_transform. This is critical. We are applying the same rules that the scaler learned from the training data to our test data. This prevents "data leakage," ensuring our test set remains a truly unseen evaluation set

#### Phase 3: Building a Baseline Model.

```python
from sklearn.linear_model import LogisticRegression
```

```python
# create an instance of the model
log_model = LogisticRegression()
# train the model on the training data
log_model.fit(x_trained_scaled, y_train)
#make prediction on scaled data
y_pred = log_model.predict(x_test_scaled)
```

### Evaluate Model Performance

```python
from sklearn.metrics import classification_report , confusion_matrix
```

* classification_report: This function builds a neat text report showing the main classification metrics.
* confusion_matrix: This function creates a table that visualizes the performance of the model by showing what it got right and what it got wrong.

```python
print('Classification Report:')
print(classification_report(y_test, y_pred))

print('Confusion Matrix')
print(confusion_matrix(y_test, y_pred))
```

#### Phase 4: Advanced Modeling and Tuning.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(random_state = 42)
rf_model.fit(x_trained_scaled, y_train)
```

```python
# Make predictions on the scaled test data
y_pred_rf = rf_model.predict(x_test_scaled)

# Generate the new classification report
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators' : [100, 200],
    'max_depth' : [10, 20],
    'class_weight' : [{0:1, 1:3},{0:1, 1:5}]
}

grid_search = GridSearchCV(estimator = RandomForestClassifier(random_state = 42),
                            param_grid = param_grid,
                            scoring = 'recall',
                            cv = 3,
                            verbose = 2)

grid_search.fit(x_trained_scaled, y_train)
```

```python
print("Best parameters found: ", grid_search.best_params_)
best_rf_model = grid_search.best_estimator_
```

```python
y_pred_tuned_rf = best_rf_model.predict(x_test_scaled)
print("\nTuned Random Forest Classification Report:")
print(classification_report(y_test, y_pred_tuned_rf))
```

### Phase 5: Model Interpretation and Deployment Concepts.

```python
# Extract feature importances from our best model
importances = best_rf_model.feature_importances_
feature_names = x_final.columns
# Create a DataFrame to view the feature importances
feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)
```

```python
# --- Visualization ---
plt.figure(figsize=(10, 8))
sns.barplot(x='importance', y='feature', data=feature_importance_df.head(15))
plt.title('Top 15 Most Important Features for Predicting Churn')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
